```python
from nltk.translate.bleu_score import corpus_bleu
from nltk.tokenize import word_tokenize

def calculate_bleu_score(reference_sentences, translated_sentences):
    """
    Calculate BLEU score for Marathi to English translation task.

    Args:
    reference_sentences (list): List of lists where each inner list contains reference translation sentences.
    translated_sentences (list): List of translated sentences.

    Returns:
    float: BLEU score.
    """
    # Tokenize reference and translated sentences
    reference_tokens = [[word_tokenize(sentence.lower())] for sentence in reference_sentences]
    translated_tokens = [word_tokenize(sentence.lower()) for sentence in translated_sentences]

    # Calculate BLEU score
    bleu_score = corpus_bleu(reference_tokens, translated_tokens)

    return bleu_score

# Example usage



bleu_score = calculate_bleu_score(ref_trans_sen, translated_sentences)
print("BLEU Score:", bleu_score)
```
